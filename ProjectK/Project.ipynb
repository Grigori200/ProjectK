```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
```

```python
data_tensor_train=data.values
survived_tensor_train=survived.values
data_tensor_test=data.values
survived_tensor_test=survived.values
```

```python
sc = StandardScaler()
data_std_train = sc.fit_transform(data_tensor_train[:, :])
```

```python
def show_classifier(model):
    model.fit(data_std_train, survived_tensor_train)
    predictions = model.predict(data_tensor_test)
    print(f'Accuracy: {accuracy_score(survived_tensor_test, predictions)}')
```

```python
knn = KNeighborsClassifier()
show_classifier(knn)
```

```python
tree = DecisionTreeClassifier()
show_classifier(tree)
```

```python
lr = LogisticRegression()
show_classifier(lr)
```

```python
nb = GaussianNB()
show_classifier(nb)
```

```python
svm = SVC()
show_classifier(svm)
```
